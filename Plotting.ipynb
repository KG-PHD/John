```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os
```

```python
os.chdir("D:\Engineering\PROJECTMARS\GSR_Data")

"""Setting Up Parameters"""
df = pd.read_csv('01_C_GSR.csv')

Time = df[' TimeSinceLoad(s)']
XShimmer = df['X (Shimmer)']
YShimmer = df['Y (Shimmer)']
ZShimmer = df['Z (Shimmer)']
SkinConductance = df['Skin Conductance']
SkinResistance = df['Skin Resistance']
PPG = df['PPG']
HRV = df['HRV']
HeartRate = df['Heart Rate']
GyroP = df['Gyro P (Shimmer)']
GyroR = df['Gyro R (Shimmer)']
GyroY = df['Gyro Y (Shimmer)']

df2 = pd.read_csv('01_C_Baseline_GSR.csv')

Time2 = df2[' TimeSinceLoad(s)']
SkinConductance2 = df2['Skin Conductance']
XShimmer2 = df2['X (Shimmer)']
YShimmer2 = df2['Y (Shimmer)']
ZShimmer2 = df2['Z (Shimmer)']
```

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2)
ax1.set_title('Skin Conductance')
ax1.plot(Time,SkinConductance, label='Raw Data',linewidth = 4)
ax1.legend(loc="upper right")


ax2.plot(Time2,SkinConductance2, label='Raw Data',linewidth = 4)
ax2.legend(loc="upper right")

ax3.plot(Time,XShimmer,linewidth = 2)
ax3.plot(Time,YShimmer,linewidth = 2)
ax3.plot(Time,ZShimmer,linewidth = 2)


ax4.plot(Time2,XShimmer2,linewidth = 2)
ax4.plot(Time2,YShimmer2,linewidth = 2)
ax4.plot(Time2,ZShimmer2,linewidth = 2)


fig2,ax5 = plt.subplots()
ax5.set_title('Skin Conductance vs Time')
ax5.plot(Time,SkinConductance, label='Raw Data',linewidth = 4)
ax5.legend(loc="upper right")
plt.xlabel('Time(s)')

fig3, (ax6,ax7) = plt.subplots(2,1)
ax6.set_title('Skin Conductance vs Time')
ax6.plot(Time,SkinConductance, label='Raw Data',linewidth = 2)
ax6.legend(loc="upper right")
ax7.set_title('Movement vs Time')
ax7.plot(Time,XShimmer,label='XShimmer',linewidth = 3)
ax7.plot(Time,YShimmer,label='YShimmer',linewidth = 3)
ax7.plot(Time,ZShimmer,label='ZShimmer',linewidth = 3)
ax7.legend(loc="lower right")
plt.xlabel('Time(s)')
plt.show()

fig3, (ax8,ax9) = plt.subplots(2,1)
ax8.set_title('PPG vs Time')
ax8.plot(Time,PPG, label='Raw Data',linewidth = 2)
ax8.legend(loc="lower left")
ax9.set_title('Movement vs Time')
ax9.plot(Time,XShimmer,label='XShimmer',linewidth = 3)
ax9.plot(Time,YShimmer,label='YShimmer',linewidth = 3)
ax9.plot(Time,ZShimmer,label='ZShimmer',linewidth = 3)
ax9.legend(loc="lower right")
plt.xlabel('Time(s)')
plt.show()
```
